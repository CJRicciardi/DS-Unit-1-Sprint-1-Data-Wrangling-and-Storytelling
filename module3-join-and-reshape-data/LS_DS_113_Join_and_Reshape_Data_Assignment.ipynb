```python
# concatenate order products prior and order products training

product_orders = pd.concat([order_products__prior, order_products__train])
                           
print(product_orders.shape)
product_orders.head()
```

```python
# make popular items easy to pull up

popular = ['Banana', 'Bag of Organic Bananas', 'Organic Strawberries', 
           'Organic Baby Spinach', 'Organic Hass Avocado', 'Organic Avocado', 
           'Large Lemon', 'Strawberries', 'Limes', 'Organic Whole Milk']

popular
```

```python
#create df of only the top products from the products df

condition = ((products['product_name']=='Banana') | 
             (products['product_name']=='Bag of Organic Bananas') | 
             (products['product_name']=='Organic Strawberries') | 
             (products['product_name']=='Organic Baby Spinach') |
             (products['product_name']=='Organic Hass Avocado') |
             (products['product_name']=='Organic Avocado') |
             (products['product_name']=='Large Lemon') |
             (products['product_name']=='Strawberries') |
             (products['product_name']=='Limes') |
             (products['product_name']=='Organic Whole Milk'))

top_prod = products[condition]

print(top_prod.shape)
top_prod.head()
```

```python
#remove unnecessary columns form top pro df

top_prod = top_prod.drop(['aisle_id', 'department_id'], axis=1)

top_prod.head()
```

```python
#view product_orders table

product_orders.head()
```

```python
#remove unecessary columns from product orders df

product_orders = product_orders.drop(['add_to_cart_order', 'reordered'], axis=1)

product_orders.head()
```

```python
#create final df

df = pd.merge(product_orders, top_prod, on='product_id', how='inner')

df.head()
```

```python
import pandas as pd
import numpy as np

table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T

table2
```

```python
# table2 -> tidy
#get treatments out of the index and into a column

table2 = table2.reset_index()

table2.head()
```

```python
#melt table to tidy layout

tidy2 = table2.melt(id_vars='index',value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])

tidy2.head()
```

```python
#rename columns

tidy2 = tidy2.rename(columns={
    'index': 'trt',
    'variable': 'name',
    'value': 'result'
})

tidy2
```

```python
#remove treatment from trt column

tidy2.trt = tidy2.trt.str.replace('treatment', '')

tidy2.head()
```

```python
# tidy2 -> table2b

table2b = tidy2.pivot_table(index='trt', columns='name',values='result')
```
