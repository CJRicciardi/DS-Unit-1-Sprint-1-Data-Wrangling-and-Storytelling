```
import pandas as pd

df = pd.read_csv('bob-ross/elements-by-episode.csv')

df.head()
```

```
df = df.drop(['EPISODE', 'TITLE'], axis=1)

df.head()
```

```
import matplotlib.pyplot as plt
```

```
import pandas as pd
import numpy as np
```

```
df_sum = df.sum(axis=0)

df_sum.head()
```

```
br_df = df_sum.reset_index()
```

```

br_df = (br_df.sort_values(0, axis=0, 
                           ascending=False).reset_index().drop(['level_0'], 
                  axis=1).rename(columns={'index': 'object', 0:'count'}))

br_df.head()
```

```
br_df.shape
```

```
br_df['object'] = br_df['object'].str.replace('_', ' ')

br_df.head()
```

```
br_df.shape
```

```
br_df['percent'] = (br_df['count']) / 381

br_df.head()
```

```
br_df.tail(30)
```

```
pd.set_option('display.max_rows',70)

filt = br_df['percent']>= .011

filt.tail()
```

```
f_df = br_df[filt]

f_df.tail()
```
